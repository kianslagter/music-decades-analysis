```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# load data
files = ['1950s.csv', '1960s.csv', '1970s.csv', '1980s.csv', '1990s.csv', '2000s.csv', '2010s.csv']
dataframes = [pd.read_csv(file) for file in files]

# get genres
all_genres = pd.concat([df['top genre'] for df in dataframes]).unique()

# count genres
counts = {genre: [] for genre in all_genres}
for df in dataframes:
    genre_counts = df['top genre'].value_counts()
    for genre in all_genres:
        counts[genre].append(genre_counts.get(genre, 0))

# dataframe
df = pd.DataFrame(counts)

# create chart
ax = df.plot(kind='bar', stacked=True)

# label
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(f'{int(height)}',
                xy=(rect.get_x() + rect.get_width() / 2, rect.get_y() + height / 2),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# label
ax.set_xticks(range(len(files)))
ax.set_xticklabels(['1950s', '1960s', '1970s', '1980s', '1990s', '2000s', '2010s'])
plt.xlabel('Decade')
plt.ylabel('Count')
plt.title('Top Genres for Each Decade')
plt.legend().remove()
    
# show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# load data
files = ['1950s2.csv', '1960s2.csv', '1970s2.csv', '1980s2.csv', '1990s2.csv', '2000s2.csv', '2010s2.csv']
dataframes = [pd.read_csv(file) for file in files]

# get genres
all_genres = pd.concat([df['top genre'] for df in dataframes]).unique()

# create directory
counts = {genre: [] for genre in all_genres}

# count genres
for df in dataframes:
    genre_counts = df['top genre'].value_counts()
    for genre in all_genres:
        counts[genre].append(genre_counts.get(genre, 0))

# create dataframe
df = pd.DataFrame(counts)

# create graph
genres_to_plot = ['rock', 'hip hop', 'pop', 'soul', 'adult standards']
for genre in genres_to_plot:
    plt.plot(df[genre], label=genre)

# labels
plt.xticks(range(len(files)), ['1950s', '1960s', '1970s', '1980s', '1990s', '2000s', '2010s'])
plt.xlabel('Decade')
plt.ylabel('Count')
plt.title('Change in Genre Popularity Over Decades')

# legend
plt.legend(bbox_to_anchor=(1.35, 1), loc='upper right')

# show plot
plt.show()
```
